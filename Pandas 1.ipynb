```python
superstore_df[["Category", "Sales"]].groupby("Category").min("Sales")
```

```python
superstore_df[["Sub-Category", "Quantity", "Profit"]].groupby("Sub-Category").sum("Quantity")
```

```python
summary_df = superstore_df[["Sub-Category", "Quantity", "Profit", "Sales", "Discount"]].groupby("Sub-Category").agg({"Quantity" : "sum","Sales" : "max", "Profit" : "mean","Discount" : "mean"})
```

```python
summary_df
```

```python
summary_df.plot( y = "Sales", kind = "barh")
```
